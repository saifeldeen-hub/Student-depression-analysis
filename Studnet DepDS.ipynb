```python
students=df[df['Profession']=="Student"]
```

```python
pt.figure(figsize=(14, 8))
sb.countplot(data=df, x='Profession', hue='Financial Stress')
pt.title('Students to Financial Presentation')
pt.ylabel('Count')
pt.xlabel('Profession')
pt.xticks(rotation=90)
pt.tight_layout()

pt.show()
```

# Financial Stress Student

```python
pt.figure(figsize=(14, 8))
ax=sb.countplot(data=students, x='Profession', hue='Financial Stress')
pt.title('Students to Financial Presentation')
pt.ylabel('Count')
pt.xlabel('Profession')
pt.xticks(rotation=90)
pt.tight_layout()
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{int(height)}', 
                    (p.get_x() + p.get_width() / 2., height),
                    ha='center', va='bottom', fontsize=8, rotation=0)
pt.show()
```
